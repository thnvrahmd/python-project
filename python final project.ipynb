```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('myexcel.csv')
data.head()
```

```python
data['Height'] = np.random.randint(150, 181, size=len(data))
data.head()
```

```python
team_distribution = data['Team'].value_counts()
team_distribution_percentage = (team_distribution / len(data)) * 100
```

```python
position_distribution = data['Position'].value_counts()
```

```python
age_distribution = data['Age'].value_counts().sort_index()
```

```python
team_salary = data.groupby('Team')['Salary'].sum().idxmax()
position_salary = data.groupby('Position')['Salary'].sum().idxmax()
```

```python
correlation_age_salary = data['Age'].corr(data['Salary'])
```

```python
#Team distribution plot with vertical team names
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Team', order=team_distribution.index)
plt.title('Distribution of Employees Across Teams')
plt.xlabel('Team')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate the x-axis labels to be vertical
plt.show()

team_distribution_percentage.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), startangle=140)
plt.title('Percentage Split of Employees Across Teams')
plt.ylabel('')
plt.show()

```

```python
#Position distribution plot
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Position', order=position_distribution.index)
plt.title('Distribution of Employees Across Positions')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()
```

```python
#Age distribution plot with individual ages
```

```python
plt.figure(figsize=(10, 6))
age_distribution.plot(kind='bar')
plt.title('Distribution of Employees by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Keep the age labels horizontal
plt.show()
```

```python
#Salary expenditure by team
```

```python
plt.figure(figsize=(10, 6))
team_salary_exp = data.groupby('Team')['Salary'].sum().sort_values(ascending=False)
team_salary_exp.plot(kind='bar')
plt.title('Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.show()
```

```python
#Salary expenditure by position
```

```python
plt.figure(figsize=(10, 6))
position_salary_exp = data.groupby('Position')['Salary'].sum().sort_values(ascending=False)
position_salary_exp.plot(kind='bar')
plt.title('Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.show()
```

```python
#Correlation between age and salary plot
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Age', y='Salary')
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
